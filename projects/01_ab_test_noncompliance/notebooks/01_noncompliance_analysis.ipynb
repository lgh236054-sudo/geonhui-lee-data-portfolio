```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_PATH = "../data/synthetic_ab_test.csv"
df = pd.read_csv(DATA_PATH)
df.head()
```

```
# Basic assignment check
df['assigned_group'].value_counts()
```

```
part_stats = df.groupby('assigned_group')['participated'].agg(['mean', 'count'])
part_stats['se'] = np.sqrt(part_stats['mean'] * (1 - part_stats['mean']) / part_stats['count'])
part_stats['ci'] = 1.96 * part_stats['se']
part_stats
```

```
ax = part_stats['mean'].plot(kind='bar', yerr=part_stats['ci'], capsize=4, figsize=(7,5), color='#1f77b4', edgecolor='black')
plt.title("Participation Balance Check")
plt.ylabel("Participation Rate")
plt.xticks([0, 1], ['Control (A)', 'Reserved Stock (B)'], rotation=0)
plt.ylim(0, 1)

for i, v in enumerate(part_stats['mean']):
    n = part_stats['count'].iloc[i]
    plt.text(i, v + 0.02, f"{round(v,3)}\n(n={n})", ha='center')

plt.show()
```

```
df_obs = df.dropna(subset=['purchase'])

itt_stats = df_obs.groupby('assigned_group')['purchase'].agg(['mean', 'count'])
itt_stats['se'] = np.sqrt(itt_stats['mean'] * (1 - itt_stats['mean']) / itt_stats['count'])
itt_stats['ci'] = 1.96 * itt_stats['se']
itt_stats
```

```
ax = itt_stats['mean'].plot(kind='bar', yerr=itt_stats['ci'], capsize=4, figsize=(7,5), color='#1f77b4', edgecolor='black')
plt.title("ITT: Purchase rate by assigned group")
plt.ylabel("Observed Purchase Rate")
plt.xticks([0, 1], ['Control (A)', 'Reserved Stock (B)'], rotation=0)
plt.ylim(0, max(0.12, itt_stats['mean'].max() * 1.5))

for i, v in enumerate(itt_stats['mean']):
    n = itt_stats['count'].iloc[i]
    plt.text(i, v + 0.005, f"{round(v,3)}\n(n={n})", ha='center')

plt.show()
```

```
at_stats = df_obs.groupby('received_treatment')['purchase'].agg(['mean', 'count'])
at_stats['se'] = np.sqrt(at_stats['mean'] * (1 - at_stats['mean']) / at_stats['count'])
at_stats['ci'] = 1.96 * at_stats['se']
at_stats
```

```
ax = at_stats['mean'].plot(kind='bar', yerr=at_stats['ci'], capsize=4, figsize=(7,5), color='#1f77b4', edgecolor='black')
plt.title("As-Treated: Purchase rate by treatment received")
plt.ylabel("Observed Purchase Rate")
plt.xticks([0, 1], ['Not Treated\n(Did not open)', 'Treated\n(Opened Email)'], rotation=0)
plt.ylim(0, max(0.12, at_stats['mean'].max() * 1.5))

for i, v in enumerate(at_stats['mean']):
    n = at_stats['count'].iloc[i]
    plt.text(i, v + 0.005, f"{round(v,3)}\n(n={n})", ha='center')

plt.show()
```
