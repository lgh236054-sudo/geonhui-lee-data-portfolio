# Project 03: Partitioned Pricing & Heterogeneous Treatment Effects (HTE)

Synthetic A/B test examining how **price framing** affects conversion.
We compare:
- Combined (Free shipping)
- Partitioned (+Shipping)

The goal is to demonstrate a **crossover interaction** by product price tier.


```python
from pathlib import Path
import pandas as pd

PROJECT_DIR = Path.cwd().parents[0]
DATA_PATH = PROJECT_DIR / "data" / "synthetic_pricing_data.csv"

if not DATA_PATH.exists():
    !python ../run_pricing_experiment.py

df = pd.read_csv(DATA_PATH)
df.head()
```

## ATE
Overall conversion difference by pricing frame.


```python
df.groupby('pricing_frame')['conversion'].mean()
```

## HTE Interaction
Conversion rates by product tier and pricing frame.


```python
df.groupby(['product_tier', 'pricing_frame'])['conversion'].mean().unstack()
```
